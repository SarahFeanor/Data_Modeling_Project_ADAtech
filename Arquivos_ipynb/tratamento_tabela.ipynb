## üíª Deletando as informa√ß√µes que n√£o queremos, criando novas tabelas e um novo arquivo

- Desenvolvemos um c√≥digo que processa a coluna '`TechInfos`', extraindo as informa√ß√µes desejadas e criando colunas espec√≠ficas (`configura√ß√£o`, `combust√≠vel`, `lugares`, `portas`, `porte`, `proced√™ncia` e `propuls√£o`). Posteriormente, removemos as colunas '`TechInfo`' e '`Equipamentos`'.


```python
import pandas as pd

# Carregando o arquivo CSV
TechInfos = 'tech_infos_amostra.csv'
df = pd.read_csv(TechInfos)

# Escolher o nome da coluna desejada
coluna_a_editar = 'TechInfos'
coluna_a_editar2 = 'Equipments'

# Extrair informa√ß√µes relacionadas √†s palavras-chave espec√≠ficas

df['Configura√ß√£o'] = df[coluna_a_editar].str.extract(r'Configura√ß√£o:(.*?);')
df['Combust√≠vel'] = df[coluna_a_editar].str.extract(r'Combust√≠vel:(.*?);')
df['Lugares'] = df[coluna_a_editar].str.extract(r'Lugares:(.*?);')
df['Portas'] = df[coluna_a_editar].str.extract(r'Portas:(.*?);')
df['Porte'] = df[coluna_a_editar].str.extract(r'Porte:(.*?);')
df['Proced√™ncia'] = df[coluna_a_editar].str.extract(r'Proced√™ncia:(.*?);')
df['Propuls√£o'] = df[coluna_a_editar].str.extract(r'Propuls√£o:(.*?);')

# Excluir coluna original
df.drop(columns=[coluna_a_editar], inplace=True)
df.drop(columns=[coluna_a_editar2], inplace=True)

# Renomeando as colunas
df.rename(columns={'Brands': 'Marca', 'Familys': 'Familia', 'Years': 'Ano_Modelo', 'Models': 'Modelo'}, inplace=True)

# Exibir as informa√ß√µes resultantes
print("\nDataFrame com informa√ß√µes mantidas:")
display(df)

tech_info_novo = 'tabela_tratada.csv'
```

# üíª Explica√ß√£o do script

- Este c√≥digo em Python utiliza a biblioteca pandas para manipular e extrair informa√ß√µes da tabela '`tech_infos_amostra.csv`'.

- Apos definir quais colunas seriam editadas, o c√≥digo extraiu informa√ß√µes espec√≠ficas da coluna escolhida ('`TechInfos`') e criou novas colunas no DataFrame para cada informa√ß√£o extra√≠da.

- Depois removeu as colunas originais que continham informa√ß√µes agora armazenadas em colunas separadas `TechInfos` e `Equipaments`.

- Houve altera√ß√£o dos nomes das colunas "`Brands`", "`Familys`", "`Years`" e "`Models`" que estavam em lingua estrangeira para "`Marca`", "`Familia`", "`Ano_modelo`" e "`Modelo`".

- Exibindo assim o DataFrame resultante, agora com colunas adicionais derivadas da coluna original '`TechInfos`'.
